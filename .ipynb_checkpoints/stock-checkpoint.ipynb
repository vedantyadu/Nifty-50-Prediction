```python
import pandas as pd

df = pd.read_csv('nifty_50.csv')
df = df.dropna()
df = df[['Date', 'Open', 'High', 'Low', 'Close']]
```

```python
from dateutil.parser import parse

def prevHoliday(prev, cur):
    dt1 = parse(prev)
    dt2 = parse(cur)
    delta = dt2 - dt1
    return 1 if delta.days > 1 else 0

size = df['Date'].size
df['After holiday'] = [0 if i == size - 1 else prevHoliday(df.iloc[i + 1]['Date'], df.iloc[i]['Date']) for i in range(size)]
df
```
